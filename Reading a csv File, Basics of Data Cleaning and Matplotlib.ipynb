# Matplotlib Basics

### Line Graphs

```python
x=[1,2,3]
y=[4,5,6]
#To plot
plt.plot(x,y)
plt.show()
```

```python
# To show a title,x-lable,y-lable
plt.plot(x,y)
plt.title("Basic Graph-")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
#To modify fonts use fontdict

#To change fonts use fontname
#To change fontsize use fontsize
#To make font style use fontweight

plt.plot(x,y)
plt.title("Basic Graph-", fontdict={"fontname": "Comic Sans MS"})
plt.xlabel("X-Axis",fontdict ={"fontsize":15})
plt.ylabel("Y-Axis",fontdict = {"fontweight":"bold"})

plt.show()
```

```python
a = [0,1,2,3,4,5]
b = [0,2,4,6,8,10]
plt.plot(a,b)
#To have only integer values
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
plt.show()
```

```python
#To add a legend to our graph use plt.legend() with label
#To add dots/marker in our graph
plt.plot(a,b, label = "2x",marker=".",markersize=10)
plt.legend()
plt.show()
```

```python
#To change colors of line,its size or style
plt.plot(a,b, label = "2x",color="red",linewidth=5,linestyle="--")
plt.legend()
plt.show()
```

#### To combine all the above steps in one line 

```python
#loookup shorthand notations documentation
plt.plot(a,b,"r.--",label="2x")
plt.legend()
plt.show()
```

#### To add a second line

```python
plt.plot(a,b,"r.--",label="2x")

x2 = np.arange(0,5,0.5)
#print(x2)
plt.plot(x2,x2**2,'r',label = "x^2")

plt.legend()
plt.show()
```

#### To show somekind of future prediction

```python
plt.figure(figsize=(5,3),dpi=200) #here dpi = ppi

plt.plot(a,b,"r.--",label="2x")

#print(x2)
#Plot part of graph as line
plt.plot(x2[:6],x2[:6]**2,'r',label = "x^2")
#Plot part of graph as dash line
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.legend()
plt.show()
```

#### To create multiple axes
use subplot cmd

```python
x = np.arange(-10, 11)
plt.figure(figsize=(12, 6))
plt.title('2x vs x^2')

plt.subplot(1, 2, 1)  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.legend(['X^2'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, 2*x)

plt.legend(['2x'])

plt.xlabel('X')
plt.ylabel('2 x X')

plt.show()
```

### Bar Charts

```python
#just change from plot to bar
x = ["A","B","C"]
y = [1,5,2]
plt.bar(x,y)
plt.show()
#You can add title, axes , color etc the same way
```

```python
#One new thing you can do is patterns on bar charts
bars = plt.bar(x,y)
bars[0].set_hatch("-")
bars[1].set_hatch("/")
bars[2].set_hatch("*")
```

```python
#One easy way to do above patterns is simply
bars = plt.bar(x,y)
patterns = ["-","/","*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0)) #pop(0) is used to correctly assign patterns
```
